```python
import pandas as pd
import json
```

```python
tweetList = []
```

```python
# twitter_streaming.py
with open('samsung.json') as json_file:
    data = json.load(json_file)
    
    for item in data :
        hashtags = ''
        for hashtag in item['entities']['hashtags'] :
            hashtags += hashtag['text'] + ' ';
        tweetList.append([item['id'],
                          item['created_at'],
                          item['text'],
                          item['user']['id'],
                          item['user']['name'],
                          item['user']['screen_name'],
                          hashtags,
                          item['source']
                         ])
dfStream = pd.DataFrame(tweetList, columns=['id','createdDate','text','userID', 'userName', 'userAccount', 'hashtags', 'source'])
dfStream
```

```python
# twitter_search.py
with open('love.json') as json_file:
    data = json.load(json_file)
    listTweets = data['statuses']
    
    for item in listTweets :
        hashtags = ''
        for hashtag in item['entities']['hashtags'] :
            hashtags += hashtag['text'] + ' ';
        tweetList.append([item['id'],
                          item['created_at'],
                          item['text'],
                          item['user']['id'],
                          item['user']['name'],
                          item['user']['screen_name'],
                          hashtags,
                          item['source']
                         ])
dfSearch = pd.DataFrame(tweetList, columns=['id','createdDate','text','userID', 'userName', 'userAccount', 'hashtags', 'source'])
dfSearch
```

```python
for tweet in dfSearch['text'] :
    print(tweet)
    print('\n')
```
